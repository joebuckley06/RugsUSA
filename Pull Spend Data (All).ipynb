```python
import pandas as pd
import numpy as np
import os
from datetime import datetime, timedelta
import sqlalchemy
import psycopg2
from sqlalchemy import create_engine
import json
# Custom JB methods
import Criteo_update
import Bing_update_script
import facebook_update
import impact_update
import google_adwords_update
import connexity_update
import pinterest_update

with open('Rugs_AWS_creds.json') as json_file:
    creds = json.load(json_file)
    
# Write all creds
host = creds['host']
port = creds['port']
username = creds['username']
password = creds['password']
database = creds['database']

# Postgres username, password, and database name
MYSQL_ADDRESS = host
## INSERT YOUR DB ADDRESS IF IT'S NOT ON PANOPLY
MYSQL_PORT = port

## CHANGE THIS TO YOUR PANOPLY/POSTGRES USERNAME
MYSQL_USERNAME = username

## CHANGE THIS TO YOUR PANOPLY/POSTGRES PASSWORD 
MYSQL_PASSWORD = password

## CHANGE THIS TO YOUR DATABASE NAME
MYSQL_DBNAME = database

# A long string that contains the necessary Postgres login information
MYSQL_str = ('mysql+pymysql://{username}:{password}@{ipaddress}:{port}/{dbname}'.format(username=MYSQL_USERNAME, password=MYSQL_PASSWORD, ipaddress=MYSQL_ADDRESS, port=MYSQL_PORT,dbname=MYSQL_DBNAME))

# Create the connection
cnx = create_engine(MYSQL_str, pool_recycle=3600) 
```

```python
# Impact Update
impact_update.update_AWS()
df_impact_db = pd.read_sql_query('''SELECT * FROM impact_spend;''', cnx)
df_impact_db
```

```python
# df_fb_db = pd.read_sql_query('''SELECT * FROM facebook_spend;''', cnx)
facebook_update.update_AWS()
df_fb_db = pd.read_sql_query('''SELECT * FROM facebook_spend;''', cnx)
df_fb_db
```

```python
# Update Bing in AWS
Bing_update_script.update_AWS()
df_bing_db = pd.read_sql_query('''SELECT * FROM microsoft_spend;''', cnx)
df_bing_db.tail()
```

```python
Criteo_update.update_AWS()
df_criteo_db = pd.read_sql_query('''SELECT * FROM criteo_spend;''', cnx)
df_criteo_db
```

```python
# Google Adwords Spend
google_adwords_update.update_AWS()
df_adwords_db = pd.read_sql_query('''SELECT * FROM google_spend;''', cnx)
df_adwords_db
```

```python
# Connexity Spend (from Google)
connexity_update.update_AWS()
df_connexity_db = pd.read_sql_query('''SELECT * FROM connexity_spend;''', cnx)
df_connexity_db
```

```python
# New Pinterest data

# Import latest pinterest download
pd_pinterest = pd.read_csv('BPR-14days.csv')
print(max(pd_pinterest['Date']))

pinterest_update.update_AWS_pin(pd_pinterest)
df_pinterest_db = pd.read_sql_query('''SELECT * FROM pinterest_spend;''', cnx)
df_pinterest_db
```

```python
df_combined_spend = pd.concat([df_criteo_db,
                               df_fb_db,
                               df_bing_db,
                               df_impact_db,
                               df_adwords_db,
                               df_connexity_db,
                               df_pinterest_db], 
                              axis=0, ignore_index=True, sort=True)
df_combined_spend
```

```python
# CSV to be used to create demo Data Studio Dashboard
df_combined_spend.to_csv('ThirdPartySpend_RUSA.csv',index=False)
```
